## Solution part 2

```python
import bs4 as bs
```

```python
soup = bs.BeautifulSoup(source, 'html.parser')
listings = soup.find_all('tr', {'class': 'dbaListing'})
# print(listings)
entries = []
for e in listings:
    title = e.select('td>div>a')[1].text.strip()
    location = e.select('td>ul>li')[1].text.strip()
    price = e.select('td>a')[1].text.strip().replace('.', '')
    try:
        img_url = e.select('td>div>a>div')[0]['data-original'].strip()
    except:
        img_url = "No image url"
    page_url = e.select('td>a')[0]['href'].strip()
    entry_tuple = (title, location, price, img_url, page_url)
    entries.append(entry_tuple)
    
print(entries)
```

```python
import pandas as pd
```

```python
sorted_by_price = sorted(entries, key=lambda tup: int(tup[2][:-3]))

# # print(entries[0][2][:-3])
# for entry in entries:
#     print(entry[2][:-3])
# # print(sorted_by_price)
df = pd.DataFrame(sorted_by_price, columns=['Title', 'Location', 'Price', 'img url', 'page url'])
df
```
